- anagrams: 단어 퍼즐과 관련된 데이터를 포함.
- anscombe: Anscombe's Quartet이라는 잘 알려진 데이터셋으로, 통계 시각화에 사용.
- attention: 주의(Attention) 관련 데이터.
- brain_networks: 뇌 네트워크와 관련된 데이터.
- car_crashes: 차량 사고 통계 데이터.
- diamonds: 다이아몬드의 가격과 특성 관련 데이터.
- dots: 점 데이터를 시각화하는 데 사용.
- dowjones: 다우존스 산업평균지수 관련 데이터.
- exercise: 운동과 건강 관련 데이터.
- flights: 항공편 탑승자 수 데이터.
- fmri: fMRI(뇌 영상) 데이터를 포함.
- geyser: 간헐천(geyser) 관련 데이터.
- glue: General Language Understanding Evaluation (GLUE) 데이터.
- healthexp: 건강 관련 지출 데이터.
- iris: 아이리스 꽃의 특성을 나타낸 유명한 데이터셋.
- mpg: 차량 연료 효율(Miles Per Gallon, MPG) 데이터.
- penguins: 펭귄 종과 그 특성에 대한 데이터.
- planets: 행성 데이터.
- seaice: 해빙(Sea Ice) 관련 데이터.
- taxis: 택시 운행 데이터.
- tips: 팁과 관련된 데이터 (음식점에서 발생하는 팁 등).
- titanic: 타이타닉 호 탑승객 데이터.

```python
plt.rc('font', family="NanumBarunGothic")

import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(13,7))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.stripplot(x='day', y='tip', hue = 'sex', data=tips, alpha = 0.5, ax=ax1)
sns.swarmplot(x='day', y='tip', hue = 'sex', data=tips, palette = 'Set2',alpha=0.5 ,  ax=ax2)
ax1.set_title('Strip plot 으로 표현한 날짜와 성별의 팁 관계')
ax1.set_title('Swarm plot 으로 표현한 날짜와 성별의 팁 관계')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.countplot(x='time', data=tips, ax=ax1)
sns.countplot(x='day', hue = 'day', data=tips, palette = 'Set2', ax=ax2)
ax1.set_title('Countplt 시간에 따른 팁 빈도차트')
ax2.set_title('Countplt 시간과 날짜에 따른 팁 빈도차트')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.regplot(x='total_bill', y='tip', data=tips, color='blue', scatter_kws={'s':50,'alpha':0.5}, line_kws={'linestyle':'--','color':'red'}, ax=ax1)
sns.regplot(x='total_bill', y='tip', data=tips, color='blue', scatter_kws={'s':50,'alpha':0.5}, fit_reg=False, ax=ax2)

fig.suptitle("Scatter plots")
ax1.set_title('선형회귀 ON')
ax2.set_title('선형회귀 OFF')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')

sns.pairplot(iris, hue="species", markers=["o","s","D"] )

plt.show()
```

```python
sns.boxplot(x="species",y="sepal_length", data=iris)
plt.title("꽃받침 길이와 종의 관계")
plt.show()
```

```python
sns.violinplot(x="species", y="petal_length", data=iris, hue="species", palette="muted")
plt.title("종에 따른 꽃잎의 길이 분포")
plt.show()
```

```python
fig = plt.figure(figsize=(12,3))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

sns.boxplot(x="species",y="sepal_length", data=iris, ax=ax1)
sns.violinplot(x="species", y="petal_length", data=iris, hue="species", palette="muted", ax=ax2)
sns.histplot(data=iris, x="petal_length", hue="species", kde=True, element="step", palette="Set2", ax=ax3)

ax1.set_title("꽃받침 길이와 종의 관계")
ax2.set_title("종에 따른 꽃잎의 길이 분포1")
ax3.set_title("종에 따른 꽃잎의 길이 분포2")

sns.lmplot(x='sepal_length', y='sepal_width', hue='species', data=iris, markers=["o","s","D"], height=6, aspect=1.2)
plt.title("꽃받침 길이와 너비의 관계")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
exercise = sns.load_dataset('exercise')
exercise.sample(5)
```

```python
# point plot
plt.figure(figsize=(7,4))
sns.pointplot(x='time', y='pulse', hue='kind', data=exercise, capsize=0.1)
plt.title("심박수와 시간별 운동 종류와 관계")
```

```python
# box plot
plt.figure(figsize=(7,4))
sns.boxplot(x='time', y='pulse', hue='kind', data=exercise, palette='Set2')
plt.title("심박수와 시간별 운동 종류와 관계")
```

```python
# swarm plot
plt.figure(figsize=(7,4))
sns.swarmplot(x='time', y='pulse', hue='kind', data=exercise, dodge=True, palette='Set1')
plt.title("시간과 운동 유형에 따른 개별 맥박 측정")
plt.show()
```
