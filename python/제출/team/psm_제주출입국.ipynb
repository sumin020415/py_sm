# 제주 출입국 분석

```python
import pandas as pd
from matplotlib import rc
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from prophet import Prophet
```

```python
# 윈도우용 폰트 경로 설정
path="C:\\Windows\\Fonts\\NanumGothic.ttf"
font_name = fm.FontProperties(fname=path).get_name()
rc('font',family=font_name)
plt.text(0.3,0.3,'한글')
```

```python
# csv 파일 임포트
file_path = "./ref/출입국_제주.csv"
ori = pd.read_csv(file_path,encoding="EUC-KR")
# ori.head()
ori.tail()
```

```python
ex1_data = ori.copy()
ex1_data.columns = ["airport","type1","gender","year","type2","value"]
# ex1_data.rename(columns=("",""))
ex1_data
```

```python
# 필터제작
filter1 = ex1_data['type1']=='내국인'
ex2_data = ex1_data[filter1].copy()
ex2_data.head()
```

```python
ex1_data = ex1_data.groupby('year')['value'].sum().reset_index(name='total')
ex1_data.head(10)
```

```python
ex1_data.plot(y='total', figsize=(10, 5), title="2010-2023년 제주 입출국 내국인 차트", color='green')
```

```python
ex2_data = ex1_data.copy()
ex2_data['year'] = pd.to_datetime(ex2_data['year'], format="%Y")
ex2_data = ex2_data.rename(columns={'year':'ds','total':'y'})
ex2_data.tail()
```

```python
m = Prophet()
m.fit(ex2_data)
```

```python
future = m.make_future_dataframe(periods=10, freq='YE') 
forecast = m.predict(future)
fig1 = m.plot(forecast)
plt.savefig("./chart/제주예측.png")
plt.figure(figsize=(10,15))
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread("./chart/제주예측.png")

fig , axs = plt.subplots(ncols=2,figsize=(8,3)) # 2 x 2 서브플롯 생성

axs[0].plot(ex1_data['year'], ex1_data['total'], color='r') 

axs[0].set_title("2018~2023년 제주 입국 내국인 차트")
axs[1].set_title("제주 입국자 향후5년 예상차트")

axins = fig.add_axes(axs[1])
axins.imshow(img)
plt.tight_layout()
plt.show()
```
