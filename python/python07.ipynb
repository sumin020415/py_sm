```python
# filter (함수, 이터러블) **
def positive(l):
    result = []
    for i in l:
        if i> 0:
            result.append(i)
    return result
print(positive([1,-3,2,0,-5,6]))
```

```python
# itemgetter2.py
from operator import itemgetter
name = ["jane","dave","sally"]
age = [22,32,17]
grade = ['A','B','B']
# students = []
# for n, a, g in zip(name,age,grade):
#     students.append({'name':n,'age':a,'grade':g})

students = [{'name':n,'age':a,'grade':g} for n, a, g in zip(name,age,grade) ]
print(students)

# students = [
#     {"name": "jane", "age": 22, "grade": 'A'},
#     {"name": "dave", "age": 32, "grade": 'B'},
#     {"name": "sally", "age": 17, "grade": 'B'},
# ]
result = sorted(students, key=itemgetter('age'))
print(result)
```

```python
# operator.attrgetter()
from operator import attrgetter

class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
    def pr(self):
        return self.name , self.age, self.grade
students = [
    Student('jane', 22, 'A'),
    Student('dave', 32, 'B'),
    Student('sally', 17, 'B'),
]
result = sorted(students, key=attrgetter('age'))
for r in result:
    print(r.pr())
```

## os

```python
import time
def long_tesk():
    for i in range(5):
        time.sleep(1)
        print(f"일하는 중...{i}")
```

```python
# 기본작업 5초 x 5회 = 25초
start = time.time()
print("=====Start=====")
for n in range(5):
    long_tesk()
print("=====End=====")
print(time.time() - start)
```

```python
# thread를 사용
import threading
start = time.time()
print("=====Start=====")
threads = []
for n in range(5):
    t = threading.Thread(target=long_tesk)
    threads.append(t)
for t in threads:
    t.start()
for t in threads:
    t.join()
print("=====End=====")
print(time.time() - start)
```

```python
# urllib
import urllib.request
def get_wikidocs(page):
    resource = 'https://wikidocs.net/{}'.format(page)
    with urllib.request.urlopen(resource) as s:
        with open('wikidocs_%s.html' % page, 'wb') as f:
            f.write(s.read())
get_wikidocs(1)
```
